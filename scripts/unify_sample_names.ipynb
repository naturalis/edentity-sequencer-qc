# Unify Elements vs NextSeq sample naming

```python


elements_esv = pd.read_parquet("/data/luka.lenaroto/eDentity/ITS-comparison/elements/elements_apscale/8_esv_table/elements_ESV_table.parquet.snappy")
elements_esv.head()
elements_esv.to_csv("/data/luka.lenaroto/eDentity/ITS-comparison/elements/elements_apscale/8_esv_table/elements_ESV_table.tsv.gz", index=False, sep="\t", compression="gzip")
```

```python
elements_esv.columns[:5]
```

```python
elements_colmuns = []
elements_markers = []
for col in elements_esv.columns:
    if "ITS" in col:
        if "Blank" in col:            
            #new_name = f'{col.split("_")[3]}_{col.split("_")[2]}'
            elements_colmuns.append(f'{col.split("_")[3]}')
            elements_markers.append(f'{col.split("_")[2]}')
        else:
            #new_name = f'{col.split("_")[4]}_{col.split("_")[2]}'
            elements_colmuns.append(f'{col.split("_")[4]}')
            elements_markers.append(f'{col.split("_")[2]}')
    else:
        elements_colmuns.append(col)
print(elements_colmuns[:10])

```

```python
elements_esv.columns = elements_colmuns
elements_esv.head()
```

```python
nextseq_esv = pd.read_parquet("/data/luka.lenaroto/eDentity/ITS-comparison/NextSeq/NextSeq_apscale/8_esv_table/NextSeq_ESV_table.parquet.snappy")
nextseq_esv.to_csv("/data/luka.lenaroto/eDentity/ITS-comparison/NextSeq/NextSeq_apscale/8_esv_table/NextSeq_ESV_table.tsv.gz", index=False, sep="\t", compression="gzip")
nextseq_esv.head()
```

```python
nextseq_columns= []
nextseq_markers = []
for col in nextseq_esv.columns:
    
    if "ITS" in col:
        if "Blank" in col:
            # new_name = f'{col.split("_")[1]}_{col.split("_")[0]}'
            nextseq_columns.append(f'{col.split("_")[1]}')
            nextseq_markers.append(f'{col.split("_")[0]}')
        else:
            #new_name = f'{col.split("_")[2]}_{col.split("_")[0]}'
            nextseq_columns.append(f'{col.split("_")[2]}')
            nextseq_markers.append(f'{col.split("_")[0]}')
    else:
        nextseq_columns.append(col)
print(nextseq_columns[:10], nextseq_markers[:10])
```

```python
nextseq_esv.columns = nextseq_columns
nextseq_esv.head()
```
